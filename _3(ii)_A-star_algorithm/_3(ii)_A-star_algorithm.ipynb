```python
def astar(start_node , stop_node):
  open_set = set(start_node)
  closed_set = set()
  g = {}
  parents = {}
  cost = 0
  g[start_node] = 0
  parents[start_node] = start_node
  while len(open_set) > 0:
    n = None
    for v in open_set:
      if (n == None) or (g[v]+heuristic[v] < g[n]+heuristic[n]):
        n = v
    if n == stop_node or Graph_nodes[n] == None:
      pass
    else:
      for (m,weight) in get_neighbours(n):
        if m not in open_set and m not in closed_set:
          open_set.add(m)
          parents[m] = n
          g[m] = g[n] + weight
        else:
          if g[m] > g[n]+weight :
            g[m] = g[n]+weight
            parents[m] = n
            if m in closed_set:
              closed_set.remove(m)
              open_set.add(m)
    if n == None:
      print("Path does not exist.")
      return None
    if n == stop_node:
      pc = g[n]
      path = []
      while parents[n] != n:
        path.append(n)
        n = parents[n]
      path.append(start_node)
      path.reverse()
      print("Path found: {}".format(path))
      print("Path cost = ",pc)
      return path
    open_set.remove(n)
    closed_set.add(n)
  print("Path does not exist")
  return None

def get_neighbours(v):
  if v in Graph_nodes:
    return Graph_nodes[v]
  else:
    return None

def heuristic(n):
  return (heuristic[n])

def add_edge(v1 , v2 , cost):
  if v1 not in Graph_nodes:
    print(v1," is not present.")
  elif v2 not in Graph_nodes:
    print(v2," is not present.")
  Graph_nodes[v1].append((v2 , cost))
  Graph_nodes[v2].append((v1 , cost))

def add_node(v):
  if(v in Graph_nodes):
    print(v," is present.")
  else:
    Graph_nodes[v] = []
    heuristic[v] = []

def getval(n,val):
  if n in Graph_nodes:
    heuristic[n] = val
  else:
    return None

# Main body
Graph_nodes = {}
heuristic = {}
v = int(input("Enter the number of vertices : "))
for i in range(v):
  vertex = input("Enter the vertex : ")
  add_node(vertex)
  heur = int(input("Enter the heuristic value of the vertex : "))
  getval(vertex , heur)
print("Creating the graph.)")
for j in range(v):
  p = input("Enter a vertex : ")
  x = int(input("Enter the number of edges from this vertex : "))
  for k in range(x):
    q = input("Enter the edge vertex : ")
    cost = int(input("Enter cost of the edge : "))
    add_edge(p,q,cost)

start = input("Enter the start/source vertex : ")
goal = input("Enter the goal/target vertex : ")
if goal not in Graph_nodes:
  print("Goal node not present.")
else:
  astar(start , goal)

```
